```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
weather_report = pd.read_csv('airport_weather.csv')
weather_report
```

```python
weather_report['NAME'].nunique()
```

```python
cities = pd.read_csv('airports_list.csv')
cities
```

```python
weather = pd.merge(cities, weather_report, how='left', on='NAME')
weather
```

```python
weather_important = weather[['DATE', 'PRCP', 'SNOW', 'SNWD', 'TMAX', 'AWND', 'ORIGIN_AIRPORT_ID']]
```

```python
weather_important.isna().sum()
```

```python
weather_important.loc[weather_important['TMAX'].isna()]
```

```python
weather_important.drop(weather_important.loc[weather_important['ORIGIN_AIRPORT_ID'].isna()].index, axis=0, inplace=True)
```

```python
weather_important.loc[weather_important['TMAX'].isna()]
```

```python
weather_important['TMAX'].fillna(round(weather_important.groupby('ORIGIN_AIRPORT_ID')['TMAX'].transform('mean'), 1), inplace=True)
weather_important['AWND'].fillna(round(weather_important.groupby('ORIGIN_AIRPORT_ID')['AWND'].transform('mean'), 1), inplace=True)
weather_important.fillna(0, inplace=True)
```
