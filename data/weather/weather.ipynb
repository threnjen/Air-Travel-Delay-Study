```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
weather1 = pd.read_csv('2682099.csv')
weather2 = pd.read_csv('2682105.csv')
weather3 = pd.read_csv('2682113.csv')
weather4 = pd.read_csv('2682125.csv')
```

```python
weather_report = pd.concat([weather1, weather2, weather3, weather4])
```

```python
weather_report
```

```python
weather_report['NAME'].nunique()
```

```python
cities = pd.read_csv('cities.csv')
cities
```

```python
weather = pd.merge(weather_report, cities, how='left', on='NAME')
weather
```

```python
weather.isna().sum()
```

```python
weather_important = weather[['DATE', 'PRCP', 'SNOW', 'SNWD', 'TMAX', 'TMIN', 'AWND', 'ORIGIN_CITY_NAME']]
```

```python
weather_important.isna().sum()
```

```python
weather_important.fillna(0, inplace=True)
```
